<h4>2.2.1 넘파이 ndarray, 리스트, 딕셔너리 -> DataFrame으로 변환</h4>
- DataFrame은 넘파이 ndarray, 리스트와 다른 칼럼명을 가짐, 이로인해 다른것 보다 편리하게 데이터 변환가능<br>
- DataFrame은 2차원 데이터임
- 2차원 이하의 데이터들만 DataFrame으로 변환가능

** DataFrame으로 변환 시 칼럼명 지정 (지정x: 자동으로 할당됨)<br>
    => data: 리스트, 딕셔너리, 넘파이 ndarray를 입력 받음<br>
    => columns: 칼럼명 리스틀 입력받아 DataFrame 생성

<h5>(1) 1차원 형태의 데이터를 기반으로 DataFrame 생성</h5> 

```python
import numpy as np
import pandas as pd

col_name1=['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# list를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 만든 DataFrame
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

<h5>(2) 2차원 형태의 데이터를 기반으로 한 DataFrame 생성</h5>
- 2x3형태의 리스트와 ndarray를 기반으로 DataFrame생성<br>
&nbsp;&nbsp;=> 칼럼명 3개 필요

```python
import numpy as np
import pandas as pd

col_name2 = ['col1', 'col2', 'col3']

list2 = [
    [1, 2, 3],
    [11, 12, 13]
]

array2 = np.array(list2)
print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

<h5>(3) 딕셔너리 -> DataFrmae 변환</h5>
- 딕셔너리 키(key) = 칼럼명 / 문자열 구성<br>
- 딕셔너리 값(value) = 키에 해당하는 "칼럼 데이터" / 리스트(ndarray)구성

```python
import pandas as pd

# key = 문자열로 칼러명 / value = 이스트 형 (또는 ndarray) 칼럼 데이터
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

<h4>2.2.2 DataFrame -> Numpy ndarray, 리스트, 딕셔너리로 변환</h4>

- DataFrame을 넘파이 ndarray로 변환 => DataFrame 객체의 values 이용하기 용이<br>
- 머신러닝 기본 패키지 기본데이터 형으로 넘파이 ndarray 사용, 데이터 핸들링으로 DataFrame을 사용하더라도 넘파이 ndarray로의 변환이 중요함

<h5>(1) DataFrame -> Numpy ndarray 변환</h5>

```python
import pandas as pd

dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)

# DataFrame -> ndarray 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
```

<h5>(2) DataFrame -> 리스트, 딕셔너리 변환</h5>

- [1] 리스트 변환  =  df_dict.values.tolist()<br>
&nbsp;&nbsp;&nbsp;&nbsp;values로 얻은 ndarray에 tolist()호출

- [2] 딕셔너리 변환  =  df_dict.to_dict('list')<br>
&nbsp;&nbsp;&nbsp;&nbsp;DataFrame객체의 to_doct()메서드 호출, 인자로 'list' 입력시 딕셔너리값이 리스트형으로 반환

```python
import pandas as pd

dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)

# DataFrame -> 리스트 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame -> 딕셔너리 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

<h4>2.5.2 DataFrame iloc[] 연산자</h4>

- 판다스는 DataFrmae의 로우, 칼럼 지정 후 데이터 선택이 가능한 인덱싱 방식으로 iloc[], loc[] 사용

(1) iloc[] (위치 기반 인덱싱) : <br>
- 세로축 : 행과 열 위치 = 0 출발점 지정
- 가로축 : 좌표 정숫값 지정<br>
** 정숫값, 정수형의 슬라이싱, 팬시 리스트 값 입력 필요
- 불린 인덱스 제공 x : 명확한 위치 기반 인덱싱이 필요하므로

(2) loc[] (명칭 기반 인덱싱): 
- 행 위치 : DataFrame의 인덱스 값
- 열 위치 : 칼럼 명칭

```python
import pandas as pd

data = {
    'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
    'Year': [2011, 2016, 2015, 2015],
    'Gender': ['Male', 'Female', 'Male', 'Male']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])

data_df
```

```python
import pandas as pd

data = {
    'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
    'Year': [2011, 2016, 2015, 2015],
    'Gender': ['Male', 'Female', 'Male', 'Male']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])

data_df

print('### iloc[] 사용해 데이터 추출')
data_df.iloc[0,0]
```

** 에러 발생

- iloc[]에 DataFrame의 인덱스 값이나 칼러명 입력시 에러 발생

ex) iloc['칼럼명'] 입력 시 발생 에러 예제

```python
import pandas as pd

data = {
    'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
    'Year': [2011, 2016, 2015, 2015],
    'Gender': ['Male', 'Female', 'Male', 'Male']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])

data_df

data_df.iloc[0, 'Name']
```

<h5>(1) iloc[] 연산 유형</h5>

iloc[1,0]           => 두번째 행, 첫번째 열 : 'Eunkyung'

iloc[2,1]           => 세번째 행, 두번째열 : '2015'

iloc[ 0:2, [0,1] ]  => 0:2 슬라이싱 범위내 1-2행, 1-2열 DataFrame 반환

(예시)

                                Name        Year
                        one     Chulmin     2011
                        two     Eunkyung    2016

iloc[0:2, 0:3]      => 0:2 범위 내 1-2 행 / 0:3 범위 내 1-3 열의 DataFrame 반환

(예시)

                                Name        Year   Gender
                        one     Chulmin     2011   Male
                        two     Eunkyung    2016   Female

iloc[:]             => 전체 DataFrame 반환

iloc[:, :]          => 전체 DataFrame 반환


** iloc[:, -1]
- 열 위치에 '-1' 입력 : DataFrame의 가장 마지막 열 값 가져오기


** iloc[:, :-1]
- 처음부터 맨 마지막 칼럼을 제외한 모든 칼럼의 값 = 피쳐 값

```python
import pandas as pd

data = {
    'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
    'Year': [2011, 2016, 2015, 2015],
    'Gender': ['Male', 'Female', 'Male', 'Male']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])

print("\n 맨 마지막 칼럼 데이터 [:, -1] \n", data_df.iloc[:, -1])
print("\n 맨 마지막 칼럼을 제외한 모든 데이터 [:, :-1] \n", data_df.iloc[:, :-1])

```

<h4>2.5.3 DataFrame loc[] 연산자</h4>

- 명칭(label) 기반 데이터 추출
- 행 : DataFrame 인덱스 값
- 열 : 칼럼명
- 형태 예시 : loc['인덱스값', '칼럼명']

```python
import pandas as pd

data = {
    'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
    'Year': [2011, 2016, 2015, 2015],
    'Gender': ['Male', 'Female', 'Male', 'Male']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])

data_df.loc['one', 'Name']
```
